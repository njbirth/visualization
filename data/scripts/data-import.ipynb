```python
import json

with open("MDB_STAMMDATEN.json", "r") as f:
    data = json.load(f);

personen = data["DOCUMENT"]["MDB"]
personen19 = []
for p in personen:
    if type(p["WAHLPERIODEN"]["WAHLPERIODE"]) == list:
        for wp in p["WAHLPERIODEN"]["WAHLPERIODE"]:
            if wp["WP"] == "19":
                personen19.append(p)
    else:
        if p["WAHLPERIODEN"]["WAHLPERIODE"]["WP"] == "19":
            personen19.append(p)

print(len(personen19))

for p in personen19:
    name = p["NAMEN"]["NAME"]

    bio = p["BIOGRAFISCHE_ANGABEN"]
    geburtsdatum = bio["GEBURTSDATUM"]
    geburtsort = bio["GEBURTSORT"]
    sterbedatum = bio["STERBEDATUM"] if bio["STERBEDATUM"] != {'-self-closing': 'true'} else None
    partei = bio["PARTEI_KURZ"]
    geschlecht = bio["GESCHLECHT"]
    familienstand = bio["FAMILIENSTAND"]
    religion = bio["RELIGION"]
    beruf = bio["BERUF"]
    md_start = None
    md_end = None
    if type(p["WAHLPERIODEN"]["WAHLPERIODE"]) == list:
        for wp in p["WAHLPERIODEN"]["WAHLPERIODE"]:
            if wp["WP"] == "19":
                md_start = wp["MDBWP_VON"]
                md_END = wp["MDBWP_BIS"] if wp["MDBWP_BIS"] != {'-self-closing': 'true'} else None
    else:
        wp = p["WAHLPERIODEN"]["WAHLPERIODE"]
        if wp["WP"] == "19":
            md_start = wp["MDBWP_VON"]
            md_END = wp["MDBWP_BIS"] if wp["MDBWP_BIS"] != {'-self-closing': 'true'} else None

    if type(name) == list:
        name = name[-1]
    print(f"{name['NACHNAME']}; {name['VORNAME']}; {geburtsdatum}; {geburtsort}; {sterbedatum}; {geschlecht}; {partei}; {familienstand}; {religion}; {beruf}; {md_start}; {md_end}")
```
